```python
import plotly.express as px

def word_freq(df):
    word_counts = df["Word"].value_counts().reset_index()
    word_counts.columns = ["Word", "Frequency"]

    top20 = word_counts.head(20)

    fig = px.bar(
        top20,
        x="Frequency",
        y="Word",
        orientation="h",
        title="Crossword Word Frequencies",
        height=500
    )
    fig.update_layout(yaxis=dict(autorange="reversed"))

    fig.show()

def word_length_dist(df):
    df["Length"] = df["Word"].str.len()

    length_counts = df["Length"].value_counts().sort_index().reset_index()
    length_counts.columns = ["Length", "Count"]

    fig = px.bar(
        length_counts,
        x="Length",
        y="Count",
        title="Word Length Distribution in Cavalier Daily Crosswords",
        labels={"Length": "Word Length", "Count": "Number of Words"},
        height=500,
        width=900
    )

    fig.show()

def words_max_length(df):
    max_len = df["Length"].max()

    longest_words = df[df["Length"] == max_len]["Word"].value_counts().reset_index()
    longest_words.columns = ["Word", "Frequency"]

    fig = px.bar(
        longest_words,
        x="Frequency",
        y="Word",
        orientation="h",
        title=f"Longest Words in Crosswords (Length = {max_len})",
        height=500
    )

    fig.update_layout(yaxis=dict(autorange="reversed"))

    fig.show()

def letter_freq(df):

    letters = "".join(df["Word"].dropna())
    letter_df = pd.Series(list(letters)).value_counts().reset_index()
    letter_df.columns = ["Letter", "Frequency"]

    fig = px.bar(letter_df, x="Letter", y="Frequency", title="Letter Frequency in Crosswords", width = 1000)

    fig.show()
    return letter_df
```

```python
import pandas as pd
import plotly.express as px

big_df = pd.read_csv("bigs.csv")

word_length_dist(big_df)

word_freq(big_df)

words_max_length(big_df)

ld = letter_freq(big_df)

print(ld.loc[ld["Frequency"]>2000])

big_df.loc[big_df["Length"]==15]
```
